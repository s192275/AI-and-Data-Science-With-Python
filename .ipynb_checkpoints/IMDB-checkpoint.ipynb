```python
from keras.datasets import imdb #keras.datasets ten imdb datası import edildi
```

```python
(train_data,train_labels),(test_data, test_labels) = imdb.load_data(num_words = 10000) #Eğitim için train ve test datasetleri 
#oluşturuldu.Ayrıca iyi yorum belirten 1 label ı ile kötü yorum belirten 0 label ı da tutulsun diye train_labels ve test_labels
#oluşturuldu.
```

```python
word_index = imdb.get_word_index() # word_index değişkeni ile imdb datasından her bir kelime(yukarıda integerlarla gösterilen ifadeler aslında kelime olarak yorumdur.) alındı.
```

```python
reverse_word_index = dict([(value,key) for (key,value) in word_index.items()]) # indexleri gerçek anlamda inceleme sözcüklerine
#dönüştüren reverse_word_index değişkeni oluşturuldu.
```

```python
decoded_review = ' '.join([reverse_word_index.get(i-3, '?') for i in train_data[1]]) # alınan kelime gruplarının aralarına boş
#luk koyarak birleştiren decoded_rewiew değişkeni oluşturuldu.
```

```python
decoded_review #oluşturulan değişken incelendi
```

```python
#içerisinde sayı bulunan listeleri sinir ağına gönderemeyiz. Dolayısıyla onları tensörlere yani matrislere çevirmemiz gerekir.
#Bunun için bir fonsiyon oluşturduk.
import numpy as np #tensör işlemi yapacağımız için numpy kullandık.
def vectorize_sequences(sequences, dimension = 10000): #fonksiyonumuzu oluşturduk.
    results = np.zeros((len(sequences), dimension)) #bir result değişkeni oluşturduk bu değişken train ve test_data daki her bir 
    #nümerik liste için boyut uzunluğu ve kelime uzunluğunu alır.
    for i, sequence in enumerate(sequences):
        results[i,sequence] = 1. #Kelime uzunluğu 1 olur
    return results        
x_train = vectorize_sequences(train_data)
x_test = vectorize_sequences(test_data)
```

```python
x_train 
```

```python
x_train[0]
```

```python
y_train = np.asarray(train_labels).astype('float32')
y_test = np.asarray(test_labels).astype('float32')
```

```python
from keras import models
from keras import layers
```

```python
model = models.Sequential()
model.add(layers.Dense(16, activation='relu',input_shape=(10000, )))
model.add(layers.Dense(16, activation= 'relu'))
model.add(layers.Dense(1, activation = 'sigmoid'))
```

```python
model.compile(optimizer='rmsprop',
              loss = 'binary_crossentropy',
              metrics = ['accuracy'])
```

```python
#Doğrulama Veri Seti Oluşturmak
x_val = x_train[:10000]
partial_x_train = x_train[10000:]

y_val = y_train[:10000]
partial_y_train = y_train[10000:]
```

```python
#Modelimizi Eğitmek
history = model.fit(partial_x_train,
                   partial_y_train,
                   epochs = 20,
                   batch_size = 512,
                   validation_data = (x_val, y_val))
```

```python
history_dict = history.history
history_dict.keys()
```

```python
loss_values = history_dict['loss']
val_loss_values = history_dict['val_loss']
epochs = range(1, len(loss_values) + 1)
import matplotlib.pyplot as plt
plt.plot(epochs, loss_values, 'bo', label = "Eğitim Kaybı")
plt.plot(epochs, val_loss_values, 'b', label = "Doğruluk Kaybı")
plt.title("Eğitim Ve Doğruluk Kaybı")
plt.xlabel("Epoklar")
plt.ylabel("Kayıp")
plt.legend()
plt.show()
```

```python
acc = history_dict['binary_accuracy']
val_acc = history_dict['val_binary_accuracy']
plt.plot(epochs, acc, 'bo', label = 'Eğitim Başarımı')
plt.plot(epochs, val_acc, 'b', label = 'DOğruluk Başarımı')
plt.title('Eğitim Ve Doğruluk Başarımı')
plt.xlabel('Epoklar')
plt.ylabel('Başarım')
plt.legend()
plt.plot()
```
